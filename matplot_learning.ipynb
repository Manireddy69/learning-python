```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels =  ['X', 'Y','Z']
values = [2,4,6]
df = plt.bar(labels, values)
patterns = ['.','*','/']
for bars in df:
    bars.set_hatch(patterns.pop(0))
#if we use pop it will arrange patterns frm start
#if we don't use pop(0) it will arrange patterns reverse

# df[0].set_hatch('.')
# df[1].set_hatch('*')
# df[2].set_hatch('/')

plt.show()
```
